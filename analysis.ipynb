# Predicting the outcome of CS duels using data from 0.5 - 3.5 seconds prior to the duel.
#### Processing the `.dem` files using the demoparser library and extracting useful features using a homemade script.

```python
import pandas as pd
import numpy as np
from demoparser2 import DemoParser
frames = []

def displacement_calc(df):
    x_change = df['X'].iloc[0] - df['X'].iloc[-1]
    y_change = df['Y'].iloc[0] - df['Y'].iloc[-1]
    return np.sqrt(x_change ** 2 + y_change ** 2)

def vector_diff(v1, v2):
    v1 = v1 / np.linalg.norm(v1)
    v2 = v2 / np.linalg.norm(v2)
    dot_product = np.dot(v1, v2)
    angle = np.arccos(dot_product)
    angle_degrees = np.degrees(angle)
    return angle_degrees

def feature_generation(df, df2):

    health = df['health'].iloc[-1]
    health_change = df['health'].iloc[0] - health
    displacement = displacement_calc(df)
    movement = df[['X','Y']].diff().abs().sum().sum()
    yaw_instability = df['yaw'].diff().abs().sum()
    weapon = df['active_weapon_name'].iloc[-1]
    helmet = df['has_helmet'].iloc[-1]
    armor = df['armor_value'].iloc[-1]
    flash_duration = df['flash_duration'].iloc[-1]

    health2 = df2['health'].iloc[-1]
    health_change2 = df2['health'].iloc[0] - health2
    displacement2 = displacement_calc(df2)
    movement2 = df2[['X','Y']].diff().abs().sum().sum()
    yaw_instability2 = df2['yaw'].diff().abs().sum()
    weapon2 = df2['active_weapon_name'].iloc[-1]
    helmet2 = df2['has_helmet'].iloc[-1]
    armor2 = df2['armor_value'].iloc[-1]
    flash_duration2 = df2['flash_duration'].iloc[-1]

    pitch = np.radians(df['pitch']).iloc[-1]
    yaw = np.radians(df['yaw']).iloc[-1]
    direction_vector = np.array([np.cos(pitch) * np.cos(yaw),
                                np.sin(pitch),
                                np.cos(pitch) * np.sin(yaw)])
    vector_0_to_1 = np.array([df2['X'].iloc[-1] - df['X'].iloc[-1], df2['Y'].iloc[-1] - df['Y'].iloc[-1], df2['Z'].iloc[-1] - df['Z'].iloc[-1]])
    offset = vector_diff(direction_vector,vector_0_to_1)

    pitch2 = np.radians(df2['pitch']).iloc[-1]
    yaw2 = np.radians(df2['yaw']).iloc[-1]
    direction_vector1 = np.array([np.cos(pitch2) * np.cos(yaw2),
                                np.sin(pitch2),
                                np.cos(pitch2) * np.sin(yaw2)])
    vector_1_to_0 = np.array([df['X'].iloc[-1] - df2['X'].iloc[-1], df['Y'].iloc[-1] - df2['Y'].iloc[-1], df['Z'].iloc[-1] - df2['Z'].iloc[-1]])
    offset2 = vector_diff(direction_vector1,vector_1_to_0)

    return health, health_change, displacement, movement, yaw_instability, offset, weapon, helmet, armor, flash_duration, health2, health_change2, displacement2, movement2, yaw_instability2, offset2, weapon2, helmet2, armor2, flash_duration2

def preprocessing(demo):
    #Creating a table containing all the duels
    demo = DemoParser(demo)
    duels = demo.parse_event('player_death')[['attacker_name','user_name','tick']].dropna()
    cols = ['health_0', 'health_change_0',
            'displacement_0', 'movement_0', 'yaw_instability_0', 'offset_0',
            'weapon_0', 'helmet_0', 'armor_0', 'flash_duration_0',
            'health_1', 'health_change_1',
            'displacement_1', 'movement_1', 'yaw_instability_1', 'offset_1',
            'weapon_1', 'helmet_1', 'armor_1', 'flash_duration_1',
            'victor']
    #Creating two tables containing all the raw data needed from each duel, one for the winner and one for the loser.
    for index, duel in duels.iterrows():
        t = duel['tick']
        ticks = [t-224,t-160,t-96,t-32]
        victor = duel['attacker_name']
        loser = duel['user_name']
        df = demo.parse_ticks(['X','Y','Z','active_weapon_name','health','armor_value','has_helmet','is_in_reload','yaw','pitch','flash_duration'],ticks=ticks)
        try:
            df_v = df[df['name'] == victor]
            df_v
            df_l = df[df['name'] == loser]
            frames.append(pd.DataFrame([list(feature_generation(df_v,df_l)+ (0,))], columns=cols))
            frames.append(pd.DataFrame([list(feature_generation(df_l,df_v)+ (1,))], columns=cols))
        except:
            print(f'An error occured in duel: {victor} Vs. {loser}')
```

#### Building a model using `xgboost` and five-fold randomised grid search hyper parameter tuning.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

#df = pd.read_csv('katowice.csv')
df = pd.concat(frames, ignore_index=True)
df['weapon_0'] = df['weapon_0'].astype('category').cat.codes
df['weapon_1'] = df['weapon_1'].astype('category').cat.codes
df['helmet_0'] = df['helmet_0'].astype(int)
df['helmet_1'] = df['helmet_1'].astype(int)

X = df.drop('victor', axis=1)
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = df['victor']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

param_grid = {'n_estimators': np.arange(100, 1001, 100),
            'learning_rate': np.linspace(0.01, 0.5, 20),
            'max_depth': np.arange(3, 15, 1),
            'subsample': np.linspace(0.5, 1.0, 10),
            'colsample_bytree': np.linspace(0.5, 1.0, 10),
            'gamma': np.linspace(0, 5, 10),
            'min_child_weight': np.arange(1, 11, 1),
            'reg_alpha': np.logspace(-5, 1, 10),
            'reg_lambda': np.logspace(-5, 1, 10)}

xgb = XGBClassifier()
random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=200,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=16,
    random_state=0)
random_search.fit(X_train, y_train)

print(f'Best parameters found: {random_search.best_params_}')
print(f'Best cross-validation accuracy: {random_search.best_score_}')

best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Test set accuracy: {accuracy:.3f}')
```
